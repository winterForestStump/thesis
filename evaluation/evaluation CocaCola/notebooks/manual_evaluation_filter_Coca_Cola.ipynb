```python
import pandas as pd
import json
from pprint import pprint

# Load your dataframe
# Replace 'your_data.csv' with the actual path to your data file
df1 = pd.read_json('/content/results_Coca_Cola_filter.json')

# Initialize dictionary to store results
results = {'index': [],
           'The context is relevant to the company': [],
           'The context is relevant to the question': [],
           'The response is correct': [],
           'The response answers the question': []}

# Check if there's a previous evaluation progress saved
try:
    with open('evaluation_results_filter', 'r') as f:
        previous_results = json.load(f)
    start_index = max(previous_results['index']) + 1
    results = previous_results
    print("Resuming evaluation from the last unanswered question.")
except (FileNotFoundError, ValueError, KeyError):
    start_index = 0
    print("Starting evaluation from the beginning.")

# Function to get valid user input (1 or 0)
def get_valid_input(message):
    while True:
        try:
            user_input = int(input(message))
            if user_input not in (0, 1):
                raise ValueError("Input must be 0 or 1.")
            return user_input
        except ValueError:
            print("Invalid input. Please enter 0 or 1.")

# Iterate over each row in the dataframe
for idx, row in df1.iloc[start_index:].iterrows():
    print("\nRow Index:", idx)
    print("Question:", row['question'])
    print("Response:", row['response'])
    print("Context:")
    pprint(row['context'], width=100)

    # Ask questions and get user input
    context_relevant_to_company = get_valid_input("Is the context relevant to the company? (1 for True, 0 for False): ")
    context_relevant_to_question = get_valid_input("Is the context relevant to the question? (1 for True, 0 for False): ")
    response_correct = get_valid_input("Is the response correct? (1 for True, 0 for False): ")
    response_answers_question = get_valid_input("Does the response answer the question? (1 for True, 0 for False): ")

    # Append results to dictionary
    results['index'].append(idx)
    results['The context is relevant to the company'].append(context_relevant_to_company)
    results['The context is relevant to the question'].append(context_relevant_to_question)
    results['The response is correct'].append(response_correct)
    results['The response answers the question'].append(response_answers_question)

    # Convert results dictionary to JSON and save to a file after each row
    json_result = json.dumps(results, indent=4)
    with open('evaluation_results_filter.json', 'w') as f:
        f.write(json_result)

    print("Progress saved.")

print("\nEvaluation results saved to 'evaluation_results_filter.json'.")

```
